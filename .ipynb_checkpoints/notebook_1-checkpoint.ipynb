```python
import numpy as np
import matplotlib.pyplot as plt

def grafic_function( func, x_range, name, m ,c ):
    x = np.linspace( x_range[0], x_range[1], 500 )
    y = func(x, m, c)

    plt.plot( x, y, label=name)

    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)

    plt.legend()
    plt.title(f"Grafica de { name }")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
def lineal_function( x , m, c ):
    return (m * x) + c


grafic_function( lineal_function, (-8, 8), "lineal function", 3, 0 )
grafic_function( lineal_function, (-5, 5), "lineal function 2", 4, 2 )
```
